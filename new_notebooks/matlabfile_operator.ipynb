```python
def binary_search(arr,target,low,high):
    if low>high:
        return -1
    mid=(low+high)//2
    if arr[mid]==target:
        return "yes"
    if arr[mid]<target:
        return binary_search(arr,target,mid+1,high)
    else:
        return binary_search(arr,target,low,mid-1)
print(binary_search([1,2,3,4],3,0,3))
```

```python
import subprocess
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
from datetime import datetime
import numpy as np

def list_directory_contents(directory):
    """
    Print all files in the directory to help with debugging
    """
    print("\nDirectory contents:")
    try:
        all_files = os.listdir(directory)
        for file in all_files:
            full_path = os.path.join(directory, file)
            file_size = os.path.getsize(full_path)
            print(f"- {file} ({file_size} bytes)")
        print(f"Total files found: {len(all_files)}")
    except Exception as e:
        print(f"Error reading directory: {str(e)}")

def run_matlab_script(script_path, script_name):
    """
    Run a MATLAB script and return the result
    """
    print(f"Running MATLAB script: {script_name}...")
    result = subprocess.run(
        ['matlab', '-batch', f"addpath('{script_path}'); {script_name};"],
        capture_output=True,
        text=True
    )
    
    # Print MATLAB output for debugging
    print("\nMATLAB Output:")
    print(result.stdout)
    if result.stderr:
        print("\nMATLAB Errors:")
        print(result.stderr)
    
    return result

def plot_figures(figure_dir, save_plots=True, output_dir=None):
    """
    Plot all figures from the specified directory
    """
    print(f"\nChecking directory: {figure_dir}")
    if not os.path.exists(figure_dir):
        print(f"Error: Directory {figure_dir} does not exist!")
        return

    # List all files in directory for debugging
    list_directory_contents(figure_dir)

    # Create output directory
    if save_plots:
        if output_dir is None:
            output_dir = os.path.join(figure_dir, 'python_plots')
        os.makedirs(output_dir, exist_ok=True)
        print(f"Output directory created at: {output_dir}")

    # Look for both MATLAB's default naming and any image files
    figures = []
    for file in os.listdir(figure_dir):
        lowercase_file = file.lower()
        # Check for various possible MATLAB figure naming patterns
        if (lowercase_file.endswith(('.png', '.jpg', '.jpeg', '.fig')) and 
            (lowercase_file.startswith('figure') or 
             'fig' in lowercase_file or 
             lowercase_file.startswith('matlab_figure'))):
            figures.append(file)
    
    figures.sort()
    print(f"\nFound {len(figures)} figures: {figures}")
    
    if not figures:
        print("\nNo MATLAB figures found. Please check:")
        print("1. Is the MATLAB script saving figures correctly?")
        print("2. Are figures being saved with a different naming pattern?")
        print("3. Are figures being saved to a different directory?")
        print("\nTry adding these lines to your MATLAB script:")
        print("figure(1);")
        print("saveas(gcf, fullfile(pwd, 'figure1.png'));")
        return
    
    # Rest of the plotting code...
    n_figures = len(figures)
    n_cols = min(3, n_figures)
    n_rows = int(np.ceil(n_figures / n_cols))
    
    fig = plt.figure(figsize=(5*n_cols, 4*n_rows))
    fig.suptitle('MATLAB Generated Figures', fontsize=16)
    
    for idx, figure in enumerate(figures, 1):
        try:
            image_path = os.path.join(figure_dir, figure)
            print(f"\nProcessing image: {image_path}")
            
            img = mpimg.imread(image_path)
            ax = fig.add_subplot(n_rows, n_cols, idx)
            ax.imshow(img)
            ax.set_title(f'Figure {idx}: {figure}')
            ax.axis('off')
            
            print(f"Successfully plotted: {figure}")
            
        except Exception as e:
            print(f"Error plotting {figure}: {str(e)}")
    
    plt.tight_layout()
    
    if save_plots:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        save_path = os.path.join(output_dir, f'combined_plots_{timestamp}.png')
        print(f"\nSaving plot to: {save_path}")
        plt.savefig(save_path, bbox_inches='tight', dpi=300)
        if os.path.exists(save_path):
            print(f"Successfully saved plot to: {save_path}")
    
    plt.show()

def main():
    script_path1 = r"C:\Users\user\Desktop\Matlabscripts\main\cascade"
    script_name1 = "cascade_MIMO_signalProcessing"
    
    result1 = run_matlab_script(script_path1, script_name1)
    
    if result1.returncode != 0:
        print("Error running MATLAB script:")
        print(result1.stderr)
        return
    
    plot_figures(
        figure_dir=script_path1,
        save_plots=True,
        output_dir=None
    )

if __name__ == "__main__":
    main()
```

```python
import subprocess
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Function to execute MATLAB script
def run_matlab_script(script_path, script_name):
    print(f"Running MATLAB script {script_name}...")
    result = subprocess.run(
        ['matlab', '-batch', f"addpath('{script_path}'); {script_name};"],
        capture_output=True, text=True
    )
    if result.returncode == 0:
        print("MATLAB script executed successfully.")
    else:
        print("Error executing MATLAB script:")
        print(result.stderr)
    return result.returncode

# Function to plot figures from the specified directory
def plot_matlab_figures(figure_dir):
    figures = [f for f in os.listdir(figure_dir) if f.startswith('figure') and f.endswith('.png')]
    
    if not figures:
        print("No figures found in the directory.")
        return

    print(f"Found {len(figures)} figures. Plotting...")
    fig, axes = plt.subplots(1, len(figures), figsize=(5 * len(figures), 5))
    
    if len(figures) == 1:
        axes = [axes]  # Ensure axes is iterable for a single plot

    for ax, figure in zip(axes, figures):
        image_path = os.path.join(figure_dir, figure)
        img = mpimg.imread(image_path)
        ax.imshow(img)
        ax.set_title(figure)
        ax.axis("off")
    
    plt.tight_layout()
    plt.show()

# Paths to MATLAB scripts and figure directory
script_path1 = r"C:\Users\user\Desktop\Matlabscripts\main\cascade"
script_name1 = "cascade_MIMO_signalProcessing"
figure_dir = r"C:\Users\user\Desktop\Matlabscripts\main\cascade\output"  # Update this to the correct directory

# Run MATLAB script and plot figures
if run_matlab_script(script_path1, script_name1) == 0:
    plot_matlab_figures(figure_dir)
```
